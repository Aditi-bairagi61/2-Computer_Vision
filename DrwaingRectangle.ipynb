```python
import cv2
cap = cv2.VideoCapture(0)

#Automatically grab width and height from video feed
#(Return float  which we need to convert to integer from later on)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

x = width//2
y = height//2
#width and height adding 1/4 of width and height from the center.
w = width//4
h = height//4
while True:
    #capture frame-by-frame
    ret,frame = cap.read()
    #Draw a rectangle on stream
    cv2.rectangle(frame,(x,y),(x+w, y+h), color=(0,0,255),thickness=4)
    #Display the resulting frame 
    cv2.imshow('frame',frame)
    if cv2.waitKey(1)&0xFF == ord('q'):
        break
#when everything is done , release the capture
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
cap = cv2.VideoCapture(0)

pt1 = (0,0)
pt2 = (0,0)
topLeft_click = False
botRight_click = False

'''
This function is triggered whenever a mouse event occures in the specified window. It checks if the left mouse button is clicked (cv2.EVENT_LBUTTONDOWN)
If both corners have already been clicked ,it resets the rectangle.
It sets pt1 to the current mouse position and sets topLeft_clicked to True.
If the bottom-right corner hasn't been clikcked yet, it sets pt2 to the current mouse position
'''
def draw_rectangle(event,x,y,flags,param):
    global pt1,pt2,topLeft_clicked,botRight_clicked
    #get mouse click
    if event == cv2.EVENT_LBUTTONDOWN:
        if topLeft_clicked == True and botRight_clicked == True:
            botRight_clicked = False
            topLeft_clicked = False
            pt1 = (0,0)
            pt2 = (0,0)
            
        if topLeft_clicked == False:
            pt1 = (x,y)
            topLeft_clicked = True
        elif botRight_clicked == False:
            pt2 = (x,y)
            botRight_clicked = True
    cap = cv2.VideoCapture(0)
    cv2.namedWindow('Test')
    cv2.setMouseCallback('Test',draw_rectangle)
    while True:
        ret,frame = cap.read()
        if topLeft_clicked:
            cv2.circle(frame,center=pt1,radius=5, color=(0,0,255),thickness=-1)
        if topLeft_clicked and botRight_clicked:
            cv2.rectangle(frame,pt1,pt2,(0,0,255),2)
        cv2.imshow('Test',frame)
        
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```
